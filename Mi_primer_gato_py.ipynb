```
def rotacion(angulo):
    import math
    import matplotlib.pyplot as plt
    plt.figure(figsize=(8,8))
    plt.axis([-15, 15, -15, 15])
    plt.axhline(0, color='gray')
    plt.axvline(0, color='gray')
    Points = {'A':[1,7],'B':[2,6],'C':[3,6],'D':[4,7], 'E':[4,5], 'F':[9,5], 'G':[9,8], 'H':[11,8], 'I':[11,7],
        'J':[10,7], 'K':[10,4], 'L':[8,4], 'M':[8,1], 'N':[6,1], 'O':[6,2], 'P':[7,2], 'Q':[7,3], 'R':[4,3],
        'S':[4,1], 'T':[2,1], 'U':[2,2], 'V':[3,2], 'W':[3,4], 'X':[1,4], 'Y':[2,5], 'Z':[3,5] }
    Lines = [ 'AB', 'BC', 'CD', 'DE', 'EF', 'FG', 'GH', 'HI', 'IJ', 'JK', 'KL', 'LM', 'MN', 'NO', 'OP', 'PQ',
        'QR', 'RS', 'ST', 'TU', 'UV', 'VW', 'WX', 'XA', 'YY', 'ZZ']
    #Lineal Transformation
    angulo = (angulo)*(0.017453)
    T = [ [ math.cos(angulo) , -math.sin(angulo)] , [ math.sin(angulo) , math.cos(angulo)] ]
    for i in Points:
        v=[Points[i][0],Points[i][1]]
        Points[i][0] = (T[0][0]*v[0]) + (T[0][1]*v[1])
        Points[i][1] = (T[1][0]*v[0]) + (T[1][1]*v[1])

    for line in Lines:
        P1 = line[0]
        P2 = line[1]
        x, y = [ Points[P1][0] , Points[P2][0] ] , [ Points[P1][1] , Points[P2][1] ]
        if P1!=P2:
            plt.plot(x,y,color='black')
        else:
            plt.plot(x,y, marker = 'o',color='black')
    plt.plot()
    plt.show()
```
